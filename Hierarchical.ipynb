```python
import numpy as np
import pandas as pd
from kmodes.kmodes import KModes
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.cluster.hierarchy import fcluster

# Example categorical data
data = pd.read_csv('DATA/Truck_drivers_final.csv')
sampled_data = data.sample(n=10000, random_state=42)

if sampled_data.isnull().any().any():
    print("Data contains NaN values. Handling NaNs...")
    sampled_data = sampled_data.fillna(method='ffill').fillna(method='bfill')
# Perform k-modes clustering

# Perform k-modes clustering on the sampled data
from kmodes.kmodes import KModes

km = KModes(n_clusters=3, init='Huang', n_init=5, verbose=1)
clusters = km.fit_predict(sampled_data)

print("Cluster centroids:\n", km.cluster_centroids_)
print("Cluster labels:", clusters)

# Plotting the results is the same as before
print("Cluster centroids:\n", km.cluster_centroids_)
print("Cluster labels:", clusters)

# Add cluster labels to data
sampled_data['cluster'] = clusters

# Calculate distance matrix using matching dissimilarity
def matching_dissim(a, b):
    return np.sum(a != b, axis=1)

centroids = km.cluster_centroids_
distance_matrix = np.zeros((len(sampled_data), len(sampled_data)))

for i in range(len(sampled_data)):
    for j in range(i + 1, len(sampled_data)):
        distance = matching_dissim(sampled_data.iloc[i, :-1].values, sampled_data.iloc[j, :-1].values)
        distance_matrix[i, j] = distance
        distance_matrix[j, i] = distance

# Perform hierarchical/agglomerative clustering on the distance matrix
Z = linkage(distance_matrix, method='average')

# Plot dendrogram
plt.figure(figsize=(10, 7))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample index')
plt.ylabel('Distance')
dendrogram(Z, labels=sampled_data.index.tolist(), leaf_rotation=90., leaf_font_size=10.)
plt.show()

# Form flat clusters
max_d = 1.5  # Distance threshold
flat_clusters = fcluster(Z, max_d, criterion='distance')

print("Flat cluster labels:", flat_clusters)
```
